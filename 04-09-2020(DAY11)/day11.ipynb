```python
from sklearn.datasets import load_iris
iris=load_iris()
iris
```

```python
data=iris['data']
```

```python
data=iris.data
```

```python
data
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(data[:,0],data[:,1])
```

```python
req=data[:,0:2]
```

```python
req
```

```python
from sklearn.model_selection  import train_test_split
cl_tr,cl_ts=train_test_split(req,test_size=0.3,random_state=7)
```

```python
from sklearn.cluster import KMeans
model=KMeans(n_clusters=3)
model.fit(cl_tr)

y_pred=model.predict(cl_ts)
y_pred
```

```python
#import pypolt
import matplotlib.pyplot as plt

#assing the cloumns of new_points :xs and ys
xs=cl_ts[:,0]
ys=cl_ts[:,1]

#make scatter plot of xs ans ys,using labels of define the color

plt.scatter(xs,ys, c=y_pred,alpha=0.5)

# assing the cluster centers :centeroides
centroids=model.cluster_centers_


#assing the columns of centeroids:centeroids_X,centeroids_Y

centroids_x=centroids[:,0]
centroids_y=centroids[:,1]

#make sctter plot of cneteroidsX and centrods_y

plt.scatter(centroids_x,centroids_y,marker='D',s=50)
plt.show()



```
