```python
import pandas as pd
```

```python
df = pd.read_csv("shirts.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
x = df[["Height","Weight"]]
x.head()
```

```python
# target
y = df["Size"]
y.head()
```

```python
y1 = pd.get_dummies(df["Size"])
y1
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier()
```

```python
knn.fit(x,y1["L"])
```

```python
# test the model
y_pred = knn.predict(x)
y_pred
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
accuracy_score(y1["L"],y_pred)
```

```python
confusion_matrix(y1["L"],y_pred)
```

```python
from sklearn.datasets import load_iris
```
