```python
examdata = pd.read_table("adminsheet.txt",header=None,sep=",")
examdata
```

```python
examdata.isnull().sum()
```

```python
x = examdata[[0,1]]
x.head()
```

```python
y = examdata[2]
y.head()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log = LogisticRegression()
```

```python
log.fit(x,y)
```

```python
pr = log.predict(x)
pr
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
accuracy_score(y,pr)
```

```python
confusion_matrix(y,pr)
```
