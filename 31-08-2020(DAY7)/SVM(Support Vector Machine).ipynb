```python
import pandas as pd
```

```python
cancer_data = pd.read_csv("cancer.csv")
cancer_data.head()
```

```python
cancer_data.shape
```

```python
cancer_data.columns
```

```python
import seaborn as sns
sns.pairplot(cancer_data,hue="diagnosis",vars=['radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean',
       'smoothness_mean'])
```

```python
x = cancer_data.iloc[:,1:31].values
x
```

```python
y = cancer_data.iloc[:,31].values
y
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
y1 = le.fit_transform(y)
y1
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y1,test_size=0.3,random_state=1)
```

```python
x_train
```

```python
from sklearn.svm import SVC
```

```python
s = SVC()
```

```python
s.fit(x_train,y_train)
```

```python
pred = s.predict(x_test)
pred
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
accuracy_score(y_test,pred)
```

```python
print(classification_report(y_test,pred))
```

```python
c = confusion_matrix(y_test,pred)
c
```

```python
sns.heatmap(c,annot=True)
```
